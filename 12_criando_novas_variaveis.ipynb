## Criando Novas Variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados2/aluguel_residencial2.csv', sep = ';')
```

```python
dados.head(10)
```

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados.head(10)
```

```python
dados['Valor m2'] = dados['Valor'] / dados['Area']
```

```python
dados.head(10)
```

```python
dados['Valor m2'] = dados['Valor m2'].round(2)
```

```python
dados.head(10)
```

```python
dados['Valor Bruto m2'] = dados['Valor Bruto'] / dados['Area']
```

```python
dados['Valor Bruto m2'] = dados['Valor Bruto m2'].round(2)
```

```python
dados.head(10)
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados.head(10)
```

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor m2', 'Valor Bruto', 'Valor Bruto m2']])
```

```python
del dados['Tipo Agregado']
```

```python
dados.head(10)
```

```python
dados_aux.head(10)
```

```python
del dados_aux['Valor Bruto']
```

```python
dados_aux.head(10)
```

```python
dados_aux.pop('Valor Bruto m2')
```

```python
dados_aux
```

```python
dados.drop(['Valor Bruto','Valor Bruto m2'], axis = 1, inplace = True)
```
